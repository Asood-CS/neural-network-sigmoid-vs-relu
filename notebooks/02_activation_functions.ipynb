# 02 - Activation Functions

As the choice of activation function is the independent variable in my investigation, it is necessary to understand how the properties of different activation functions lead to varied downstream effects in network performance and stability. I decided to choose the two most popular activation functions in AI development: Sigmoid and ReLU. Modeling and differentiating these functions helped me understand how Sigmoid's vanishing gradients contribute to a slower but more stable training environment, while ReLU's unbounded gradients yield more chaotic robustness.

```python
import numpy as np
import matplotlib.pyplot as plt

#Defining curve equation and setting bounds for graph
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 1000)
y = sigmoid(x)

plt.figure(figsize=(6, 5))
ax = plt.gca()
ax.plot(x, y, color='red', label='S(x)')

#Formatting the axes to cross at origin
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))

#Setting curve limits and bounds
ax.set_xlim(-10, 10)
ax.set_ylim(-0.1, 1.1)
ax.spines['left'].set_bounds(-0.1, 1.1)
ax.spines['bottom'].set_bounds(-10, 10)

#Ticks at each label
xticks = np.arange(-10, 11, 2)
yticks = np.linspace(0, 1, 6)
ax.set_xticks(xticks)
ax.set_yticks(yticks)
ax.set_xticklabels([str(v) if v != 0 else '' for v in xticks])
ax.set_yticklabels([str(round(v, 1)) if v != 0 else '' for v in yticks])

#Setting axis labels
ax.set_xlabel('x', loc='right')
ax.set_ylabel('S(x)', loc='top', rotation=0)

#Plotting graph in a downloadable format
ax.grid(False)
plt.tight_layout()
plt.savefig('Sigmoid.png', dpi=300)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

#Expressing the derivative of sigmoid in terms of the original function
def sig_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

x = np.linspace(-10, 10, 1000)
y1 = sigmoid(x)
y2 = sig_derivative(x)

#Include multiple curves plotted on one x axis
plt.figure(figsize=(6, 5))
ax = plt.gca()
ax.plot(x, y1, label='S(x)', color='red')
ax.plot(x, y2, label="S'(x)", color='blue')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))

ax.set_xlim(-10, 10)
ax.set_ylim(-0.1, 1.1)
ax.spines['left'].set_bounds(-0.1, 1.1)
ax.spines['bottom'].set_bounds(-10, 10)

xticks = np.arange(-10, 11, 2)
yticks = np.linspace(0, 1, 6)
ax.set_xticks(xticks)
ax.set_yticks(yticks)
ax.set_xticklabels([str(v) if v != 0 else '' for v in xticks])
ax.set_yticklabels([str(round(v, 1)) if v != 0 else '' for v in yticks])

ax.set_xlabel('x', loc='right')
ax.set_ylabel('y', loc='top', rotation=0)

#Creating a legend away from curve
plt.subplots_adjust(right=0.82)
ax.legend(loc='center left', bbox_to_anchor=(1.02, 0.5), borderaxespad=0.)

ax.grid(False)
plt.tight_layout()
plt.savefig('Sigmoid_plus_derivative.png', dpi=300)
plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt


#Defining curve equation and setting bounds for graph
def relu(x):
    return np.maximum(x, 0)

x = np.linspace(-10, 10, 1000)
y = relu(x)

plt.figure(figsize=(6, 5))
ax = plt.gca()
ax.plot(x, y, color='red', label='R(x)')

#Formatting the axes to cross at origin
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))

#Setting curve limits and bounds
ax.set_xlim(-10, 10)
ax.set_ylim(-1, 10)
ax.spines['left'].set_bounds(-1, 10)
ax.spines['bottom'].set_bounds(-10, 10)

#Ticks at each label
xticks = np.arange(-10, 11, 2)
yticks = np.arange(0, 11, 2)
ax.set_xticks(xticks)
ax.set_yticks(yticks)
ax.set_xticklabels([str(v) if v != 0 else '' for v in xticks])
ax.set_yticklabels([str(v) if v != 0 else '' for v in yticks])

#Setting axis labels
ax.set_xlabel('x', loc='right')
ax.set_ylabel('R(x)', loc='top', rotation=0)

#Plotting graph in a downloadable format
ax.grid(False)
plt.tight_layout()
plt.savefig('ReLU.png', dpi=300)
plt.show()
```
